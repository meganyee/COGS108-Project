# Research Question

Does playing back-to-back games in the NBA affect a player's individual performance? Specifically, does fatigue influence key performance metrics, such as points per game, rebounds, and turnovers?

## Background and Prior Work


The NBA, also known as the National Basketball Association, consists of 82 games for each team, and there are 32 teams, in a regular season from October through April. This means that some star players could be playing back-to-back games for the team to have the best chance of winning. This leads us to our research question of does playing back-to-back games affects a player’s performance, in areas such as points per game (PTS), rebounds, and turnovers. After doing some initial research about our topic, we could not find any player’s performance related to the back-to-back games, however, we could find the team’s performance related to the back-to-back games. Currently, we know that players tend to get worse on the second day of a back-to-back schedule. Thus, comparing the performance of a player when they have a one-day break to when they have a back-to-back game will allow us to get an in-depth statistic to generate a strong conclusion to the question.

Previously, a similar study was conducted to determine the margin of victory by an NBA team on the second night of a back to back game. They found that at home, teams played 1 point worse on the second night, compared to 2.5 points worse if they played away.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) While this information tells us that teams do in fact play worse on the second night of a back to back game, it’s a very small margin and it doesn’t account for this new resting idea in the modern NBA. More often in recent years, NBA teams have started to rest their star players in back to back games, whether this means less minutes or not playing them at all. On average, NBA teams play 16% of their games as a back to back and most commonly, they play 63% of their games with a one day break. As a result, a more accurate way to accommodate for the potential rest of players, the Sparese Impacts NBA Model was designed. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) This model creates a numerical point value of a player’s worth which can be compared to a team’s performance on the second night of a back to back. While this previous study gives us an insight into how teams usually perform on the second night of a back to back, we specifically want to study individual players’ performances so we can analyze player consistency.

The second study we found relating to the topic examines how back-to-back games impact NBA team performance, specifically when travel is involved. Using data from 2014-2021, the study found that teams generally perform worse on the second game of a back-to-back, with an average decline in margin of victory by 2.5 points. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Statistical analysis through Welch’s t-test confirmed a significant difference in performance, where “mean margin of victory on a back-to-back with travel” was measured at -2.33 points compared to 0.6 points without travel. This effect is even more noticeable when the teams need to travel, increasing the margin by nearly 3 points. By distinguishing games where teams travel from those where they remain at the same location, the study concludes that travel-induced fatigue may considerably impact player performance on back-to-backs, suggesting that reducing travel during these games could maintain game quality and competitiveness.

## Dataset #1 (Game Schedule Filtering)

```python
import pandas as pd
import re

files = [
    "C:/Users/Steven/COGS108/NBA - Steven.csv",
    "C:/Users/Steven/COGS108/NBA - Megan.csv",
    "C:/Users/Steven/COGS108/NBA - Junmo.csv",
    "C:/Users/Steven/COGS108/NBA - Hung.csv",
]

columns_to_keep = ["PLAYER NAME", "DATE", "MIN", "PTS", "REB", "TOV"]
dataframes = []
for file in files:
    df = pd.read_csv(file)
    df = df[columns_to_keep]
    dataframes.append(df)

combined_df = pd.concat(dataframes, ignore_index=True)
output_file = "Combined_and_Cleaned_NBA_Data.csv"
combined_df.to_csv(output_file, index=False)

combined_file = "Combined_and_Cleaned_NBA_Data.csv"
combined_data = pd.read_csv(combined_file)

def remove_day_of_week(date):
    if pd.isnull(date):
        return date
    return re.sub(r'\b(Mon|Tue|Wed|Thu|Fri|Sat|Sun)\b[\n\s]*', '', date, flags=re.IGNORECASE)

combined_data['DATE'] = combined_data['DATE'].apply(remove_day_of_week)
updated_file_path = "Combined_and_Cleaned_NBA_Data_DOTW_Removed.csv"
combined_data.to_csv(updated_file_path, index=False)

file_path = "Combined_and_Cleaned_NBA_Data_DOTW_Removed.csv"
data = pd.read_csv(file_path)

def add_year_to_date(date_str):
    month, day = map(int, date_str.split('/'))
    year = 2023 if month >= 10 else 2024
    return f"{month}/{day}/{year}"

data['DATE'] = data['DATE'].apply(add_year_to_date)
updated_file_path = "Combined_and_Cleaned_NBA_Data_DOTW_Removed_Add_Year.csv"
data.to_csv(updated_file_path, index=False)
```

This script takes multiple datasets of NBA player statistics from each of the group member and combines them into a single, clean dataset for analysis. It starts by selecting the desired metrics like player names, game dates, points scored, rebounds, and turnovers, and merges all the data into one file. Next, it cleans up the dates by removing unnecessary details like the day of the week (e.g., "Mon" or "Tue") and then adds missing year information. Since NBA seasons span two calendar years, the script assigns 2023 to games played from October to December and 2024 to games from January to April.

## Dataset #2 (Player Performance Metrics and Filtering)

```python
import pandas as pd

points_df = pd.read_csv('Pts_per_Game.csv', index_col=0)
turnover_df = pd.read_csv('Turnover.csv', index_col=0)
rebound_df = pd.read_csv('Rebound.csv', index_col=0)

combined_df = pd.concat([points_df, turnover_df, rebound_df], ignore_index=True)
unique_combined_df = combined_df.drop_duplicates(subset=['Player', 'TEAM'])
unique_combined_df = unique_combined_df[['Player', 'TEAM', 'PTS', 'TOV', 'REB']]
unique_combined_df.rename(columns={'Player': 'PLAYER NAME'}, inplace=True)
unique_combined_df.to_csv('Top_Players.csv', index=False)

combined_games_df = pd.read_csv('Combined_and_Cleaned_NBA_Data_DOTW_Removed_Add_Year.csv')
top_players_df = pd.read_csv('Top_Players.csv')
filtered_top_players_df = combined_games_df[combined_games_df['PLAYER NAME'].isin(top_players_df['PLAYER NAME'])]
filtered_top_players_df.to_csv('Filtered_Top_Players_Games.csv', index=False)

file_path = "Filtered_Top_Players_Games.csv"
data = pd.read_csv(file_path)

data['DATE'] = pd.to_datetime(data['DATE'])
data = data.sort_values(by=['PLAYER NAME', 'DATE'])
data['DATE_DIFF'] = data.groupby('PLAYER NAME')['DATE'].diff().dt.days
data['IS_BACK_TO_BACK'] = (data['DATE_DIFF'] == 1) | (data['DATE_DIFF'].shift(-1) == 1)
back_to_back_games = data[data['IS_BACK_TO_BACK']]
back_to_back_games = back_to_back_games.drop(columns=['DATE_DIFF', 'IS_BACK_TO_BACK'])
back_to_back_file_path = "Back_to_Back_Games.csv"
back_to_back_games.to_csv(back_to_back_file_path, index=False)

file_path = "Back_to_Back_Games.csv"
data = pd.read_csv(file_path)

data['DATE'] = pd.to_datetime(data['DATE'])
data = data.sort_values(by=['PLAYER NAME', 'DATE'])
data['DATE_DIFF'] = data.groupby('PLAYER NAME')['DATE'].diff().dt.days
second_day = data[data['DATE_DIFF'] == 1]
data['PREV_DATE_DIFF'] = data.groupby('PLAYER NAME')['DATE_DIFF'].shift(-1)
first_day = data[data['PREV_DATE_DIFF'] == 1]
first_day = first_day.drop(columns=['DATE_DIFF'], errors='ignore')
second_day = second_day.drop(columns=['DATE_DIFF', 'PREV_DATE_DIFF'], errors='ignore')

first_day_file_path = "Back_to_Back_First_Day.csv"
second_day_file_path = "Back_to_Back_Second_Day.csv"
first_day.to_csv(first_day_file_path, index=False)
second_day.to_csv(second_day_file_path, index=False)

all_games_df = pd.read_csv('Combined_and_Cleaned_NBA_Data_DOTW_Removed_Add_Year.csv')
back_to_back_df = pd.read_csv('Back_to_Back_Games.csv')

all_games_df['DATE'] = pd.to_datetime(all_games_df['DATE'])
back_to_back_df['DATE'] = pd.to_datetime(back_to_back_df['DATE'])
non_back_to_back_df = all_games_df.merge(
    back_to_back_df[['PLAYER NAME', 'DATE']],
    on=['PLAYER NAME', 'DATE'],
    how='left',
    indicator=True
)
non_back_to_back_df = non_back_to_back_df[non_back_to_back_df['_merge'] == 'left_only']
non_back_to_back_df = non_back_to_back_df.drop(columns=['_merge'])
non_back_to_back_df.to_csv('Non_Back_to_Back_Games.csv', index=False)
```

This script processes NBA game data to analyze player performance, focusing on top players and their performance in back-to-back games. It identifies top players based on points, turnovers, and rebounds, then filters game data to include only these players. Back-to-back games are flagged, separated into first and second days, and saved into separate datasets. Non-back-to-back games are also identified and stored. This preparation enables detailed analysis of performance trends under different playing conditions.

# Results

## Exploratory Data Analysis



### Section 1 of EDA - Scatterplots and Boxplots for Stats, First and Second day B2B games

We used scatterplots and boxplots to compare player performance on the first and second days of back-to-back games. The scatterplots show how points, rebounds, and turnovers on Day 1 compare to Day 2, but they’re hard to interpret because so many points overlap. To address this, we added boxplots, which make it easier to spot overall trends. From the boxplots, we see that points scored drop slightly on Day 2, while rebounds and turnovers remain about the same. This suggests fatigue might impact scoring more than other parts of the game, but further analysis is needed to confirm this.

```python
import pandas as pd
import matplotlib.pyplot as plt

first_day_df = pd.read_csv("Back_to_Back_First_Day.csv")
second_day_df = pd.read_csv("Back_to_Back_Second_Day.csv")
non_b2b_games_df = pd.read_csv("Non_Back_to_Back_Games.csv")

merged_df = pd.merge(first_day_df, second_day_df, on='PLAYER NAME', suffixes=('_Day1', '_Day2'))

metrics = ['PTS', 'REB', 'TOV']

for metric in metrics:
    plt.figure(figsize=(8, 6))
    plt.scatter(merged_df[f'{metric}_Day1'], merged_df[f'{metric}_Day2'], alpha=0.7, label='Day 1 vs Day 2')
    plt.scatter(non_b2b_games_df[metric], non_b2b_games_df[metric], alpha=0.7, color='orange', label='Non-B2B Games')
    plt.title(f'{metric}: Comparisons Between Back-to-Back and Non-B2B Games')
    plt.xlabel(f'{metric} on Day 1')
    plt.ylabel(f'{metric} on Day 2')
    plt.axline((0, 0), slope=1, color='red', linestyle='--', label='Equal Performance Line')
    plt.legend()
    plt.grid(True)
    plt.show()

plt.figure(figsize=(15, 6))
for i, metric in enumerate(metrics, 1):
    data_to_plot = [
        merged_df[f'{metric}_Day1'],
        merged_df[f'{metric}_Day2'],
        non_b2b_games_df[metric]
    ]
    plt.subplot(1, 3, i)
    plt.boxplot(data_to_plot, tick_labels=['Day 1', 'Day 2', 'Non-B2B'])
    plt.title(f'{metric}: Day 1, Day 2, and Non-B2B Comparisons')
    plt.ylabel(metric)
plt.tight_layout()
plt.show()
```

### Section 2 of EDA - Means, T-Stat, P-Value

We used the same metrics and dataset from Section 1 to conduct a paired t-test for each metric, comparing Day 1 and Day 2 of the back-to-back games. We also calculated the p-value for each metric to determine the statistical significance. Our results showed that the differences in the mean values for each stat between the two days were minor, showing that the metrics are very similar overall. However, with the p-value at the 0.05 significance level, the p-values showed that there is significant change for PTS and RBS but not for TOV.

```python
import pandas as pd
from scipy.stats import ttest_rel, ttest_ind

first_day_df = pd.read_csv("Back_to_Back_First_Day.csv")
second_day_df = pd.read_csv("Back_to_Back_Second_Day.csv")
non_b2b_games_df = pd.read_csv("Non_Back_to_Back_Games.csv")

merged_b2b_df = pd.merge(first_day_df, second_day_df, on='PLAYER NAME', suffixes=('_Day1', '_Day2'))

metrics = ['PTS', 'REB', 'TOV']

metrics_summary = {}

for metric in metrics:
    day1_mean = merged_b2b_df[f'{metric}_Day1'].mean()
    day2_mean = merged_b2b_df[f'{metric}_Day2'].mean()
    day1_variance = merged_b2b_df[f'{metric}_Day1'].var()
    day2_variance = merged_b2b_df[f'{metric}_Day2'].var()
    non_b2b_variance = non_b2b_games_df[metric].var()
    t_stat_day1_day2, p_value_day1_day2 = ttest_rel(merged_b2b_df[f'{metric}_Day1'], merged_b2b_df[f'{metric}_Day2'])
    non_b2b_mean = non_b2b_games_df[metric].mean()
    t_stat_non_b2b_day2, p_value_non_b2b_day2 = ttest_ind(
        non_b2b_games_df[metric].dropna(), merged_b2b_df[f'{metric}_Day2'], equal_var=False
    )
    metrics_summary[metric] = {
        'Day 1 Mean': day1_mean,
        'Day 2 Mean': day2_mean,
        'Non-B2B Mean': non_b2b_mean,
        'Day 1 Variance': day1_variance,
        'Day 2 Variance': day2_variance,
        'Non-B2B Variance': non_b2b_variance,
        'Difference (Day1 - Day2)': day1_mean - day2_mean,
        'Difference (Non-B2B - Day2)': non_b2b_mean - day2_mean,
        'T-Statistic (Day1 vs Day2)': t_stat_day1_day2,
        'P-Value (Day1 vs Day2)': p_value_day1_day2,
        'T-Statistic (Non-B2B vs Day2)': t_stat_non_b2b_day2,
        'P-Value (Non-B2B vs Day2)': p_value_non_b2b_day2
    }

metrics_summary_df = pd.DataFrame(metrics_summary).T
metrics_summary_df.index.name = 'Metric'

updated_file_path = "MetricSummary.csv"
metrics_summary_df.to_csv(updated_file_path)

print(metrics_summary_df)
```

# Discussion and Conclusion

Our research aimed to determine if playing back-to-back games affects NBA players' individual performance, specifically in terms of points scored (PTS), rebounds (REB), and turnovers (TOV). As an earlier hypothesis, we predicted that fatigue not only affects a player’s physical abilities but can also negatively impact their cognitive functions, which may lead to an increase in turnovers. From our analysis, we observed some significant differences that support parts of our hypothesis, while others revealed surprising results.


Firstly, the mean points scored (PTS) decreased by 0.25 points from Day 1 (18.44) to Day 2 (18.19). With the significant level at p-value = 0.05, the paired t-test showed this drop was statistically significant (p-value = 0.0011). This suggests that fatigue may play a role in reducing scoring ability, possibly because players have less energy or focus on the second day. The t-statistic (2.55) also shows that this result is meaningful, even though the difference in points is relatively small. Rebounds (REB) also showed a small but statistically significant decrease (p-value < 0.001), with the mean dropping from Day 1 (6.61) to Day 2 (6.48). The t-statistic (3.37) highlights that this decrease is meaningful despite its small magnitude, suggesting that fatigue slightly affects a player’s ability to maintain physical engagement. Lastly, the mean value for turnovers (TOV) slightly decreased from Day 1 (2.07) to Day 2 (2.04). The change wasn’t statistically significant (p-value = 0.074), and the t-statistic (1.78) shows that the difference isn’t large enough to reject the null hypothesis. This means turnovers might not be as affected by fatigue as we predicted earlier.


In conclusion, our results show that back-to-back games have some impact on individual performance, particularly in scoring and rebounding, which aligns with the hypothesis that fatigue negatively affects physical abilities. However, turnovers didn't show a significant change and even slightly decreased on Day 2, which goes against our hypothesis that fatigue would negatively impact their cognitive functions which may lead to an increase in turnovers.This suggests that while fatigue does affect physical performance metrics like scoring and rebounding, its impact on cognitive functions may not be as strong as expected. It’s also possible that factors like team strategies or how opponents play have a bigger role in turnover rates than individual fatigue.


For better comparison, we also compared non-back-to-back games to Day 2 to see how performance differs with more rest, providing additional insights into the impact of fatigue on player stats. One interesting fact that we did find in our data is that the difference between the points from a non-back-to-back game compared to a game on day 2 is actually greater than the difference between the first day of a back-to-back and the second day. Although this difference is greater, we noticed that the p-value actually increases, proving that our result is actually not statistically significant. We believe that the increase in p-value might be due to a lack of sample size in back-to-back games, compared to all the other non-back-to-back games in the NBA.


From another perspective, the changes we found, like a 0.25 points drop in scoring, are pretty small and might not have a big impact on the outcome of most games. This suggests that while fatigue is still a factor, it may not always have a major effect on individual stats. Future research could look into other metrics, like shooting percentage or defensive plays, to get a better idea of how fatigue affects different parts of a player’s performance. It would also be interesting to see if things like age, position, or how many minutes a player plays make some players more affected by back-to-back games than others. This could help teams come up with better ways to manage players’ workloads.
